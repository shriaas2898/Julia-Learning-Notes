# Packages
Packages are a huge part of Julia ecosystem there are thousands of packages for Julia and the number just keeps growing!

To see all available packages, checkout:

https://juliaobserver.com/ or https://julialang.org/packages/

We can install new packages in Julia as:

```julia
using Pkg
Pkg.add("HTTP")
```

After installing a package we use the package in current program as:

```julia
using HTTP
p = string(HTTP.request("GET", "https://raw.githubusercontent.com/JuliaRegistries/General/master/Registry.toml").body);
count(s -> occursin("{ name", s), split(p, '\n'))
```
