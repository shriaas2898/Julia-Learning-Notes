# Packages
Packages are a huge part of Julia ecosystem there are thousands of packages for Julia and the number just keeps growing!

To see all available packages, checkout:

https://juliaobserver.com/ or https://julialang.org/packages/

We can install new packages in Julia as:

```julia
using Pkg
Pkg.add("Example")
```

After installing a package we use it in any program as:

```julia
using Example
```

Example package contains a function called `hello()` which greets the name or whatver string is passed through it. Now that we have imported `Example` module we can access its functions:

```julia
hello("there! this is me here!")
```

Here is a cool Julia package:

```julia
using Colors
```

```julia
palette = distinguishable_colors(100)
```

The above function returns an array of as many distinguishable colors as we ask it to.

Then we can use these colors anywhere for example we can create a matrix using random colors from the `palette`:

```julia
rand(palette, 4, 5)
```
